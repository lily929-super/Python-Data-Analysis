# 项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

# 评估数据

## 评估数据整洁度

```python
original_data.sample(10)
```

##### 从抽样的10条数据来看，数据结构符合每列一个变量，每行一个观察值，每个单元格是一个值。具体来说每行是一个商品的一次交易记录，每列是交易相关的各个变量，不存在数据结构问题

## 评估数据干净度

```python
original_data.info()
```

##### 数据一共有541909条，Description，CustomerID存在缺失值，InvoiceDate的数据类型应为日期，CustomerID的数据类型硬是字符串，应该进行数据格式转换。

## 评估缺失数据

### 根据条件抽取出"Description"有缺失值的观察值

```python
original_data[original_data["Description"].isnull()]
```

#### 从运行结果来看，Description为空的数据有1454条 
观察这些数据，发现Description为空的同时UnitPrice=0，为了验证这个猜想，增加筛选条件，看是否存在Description为空并且nitPrice不是0的数据

```python
original_data[(original_data["Description"].isnull()) & (original_data["UnitPrice"] != 0)]
```

##### 筛选结果为0，说明缺失Description值的数据，同时也不具备有效的UnitPrice 
##### 我们要分析的是畅销商品的数据，商品名称（Description）和商品价格（UnitPrice）都是进行后续分析的重要变量，如果值缺失则不能提供有效含义，后续可以进行删除。

```python
original_data[original_data["CustomerID"].isnull()]
```

##### CustomerID是顾客编号，并不是分析畅销商品的必要变量，从CustomerID有缺失值的观察数据来看，有些交易仍然有效，因此保留此缺失值的数据

## 评估重复数据

##### InvoiceNo,StockCode,CustomerID虽然是唯一标识符，但一次交易可以有多个商品，所以InvoiceNo可以重复，不同交易可以有同一个商品，同一个顾客也可以购买不同商品，所以对于数据集无需评估重复数据

## 评估不一致数据

### 不一致数据可能存在于Country中，看看是否存在不同内容表示同一国家的情况

```python
original_data["Country"].value_counts()
```

##### 从统计结果来看，U.K. ，UK，United Kingdom都表示英国，United States，USA都是美国，清洗数据时需要对这些值进行统一

## 评估无效或错误数据

### 通过DataFrame的describe方法，对数值统计信息快速了解

```python
original_data.describe()
```

#### 筛选出Quantity是负数的观察值

```python
original_data[original_data["Quantity"] < 0]
```

##### 判断Quantity为负数的观察值，是否InvoiceNo都是以"C"开头的

```python
original_data[(original_data["Quantity"] < 0) & (original_data["InvoiceNo"].str[0] != "C")]
```

##### Quantity是负数，并且InvoiceNo不是"C"开头的观察值中，是否存在UnitPrice不为0的观察值

##### 经过上记筛选后，发现Quantity为负数的观察值满足下记两种情况 
- 发票编号(InvoiceNo)以字母"C"开头，表示订单被取消
- 商品单价(UnitPrice)是0
##### 这些交易数据不是正规的销售数据，所以不用做分析，可以把Quantity为负数的观察值删掉

#### 筛选出UnitPrice是负数的观察值

```python
original_data[original_data["UnitPrice"] < 0]
```

##### 从输出来看，Adjust bad debt是坏账调整，也不是正常的交易数据，可以删除掉

# 清洗数据

##### 根据前面评估部分得到的结论，我们需要进行的数据清理包括： 

- 把InvoiceDate变量的数据类型转换为为日期
- 把CustomerID变量的数据类型转换为字符串
-把Description变量缺失的观察值删除
- 把Country变量值"USA"替换为"United States"
- 把Country变量值"UK"、"U.K."替换为"United Kingdom"
- 把Quantity变量值为负数的观察值删除
- 把UnitPrice变量值为负数的观察值删除负数的观察值删除负数的观察值删除

##### 为了区分开经过清理的数据和原始的数据，我们创建新的变量cleaned_data，让它为original_data复制出的副本。我们之后的清理步骤都将被运用在cleaned_data上。

```python
cleaned_data = original_data.copy()
cleaned_data.head()
```

```python
cleaned_data["InvoiceDate"] = pd.to_datetime(cleaned_data["InvoiceDate"])
```

```python
cleaned_data.sample(3)
```

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].astype(str)
```

```python
cleaned_data["CustomerID"]
```

##### 把CustomerID变量值结尾的".0"删除：

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].str.slice(0,-2)
```

```python
cleaned_data["CustomerID"]
```

```python
cleaned_data.sample(2)
```

```python
cleaned_data.dropna(subset=["Description"],inplace=True)
```

```python
cleaned_data["Description"].isnull().sum()
```

```python
cleaned_data["Country"] = cleaned_data["Country"].replace("USA","United States")
```

```python
len(cleaned_data[cleaned_data["Country"] == "USA"])
```

```python
cleaned_data["Country"] = cleaned_data["Country"].replace({"UK":"United Kingdom","U.K.":"United Kingdom"})
```

```python
len(cleaned_data[(cleaned_data["Country"] == "UK") | (cleaned_data["Country"] == "U.K.")])
```

```python
cleaned_data["Country"].value_counts()
```

```python
cleaned_data = cleaned_data[cleaned_data["Quantity"] >= 0]
```

```python
len(cleaned_data[cleaned_data["Quantity"] < 0])
```

```python
cleaned_data = cleaned_data[cleaned_data["UnitPrice"] >= 0]
```

```python
len(cleaned_data[cleaned_data["UnitPrice"] < 0])
```

### 完成数据清理后，把干净整齐的数据保存到新的文件里，文件名为e_commerce_cleaned.csv。
